Conclusion:


In comparison to male users, the total number of female users is considerable.
The newest user(s) is 21 years old, while the oldest is 79.
The majority of users are between the ages of 26 and 35.



If you look at the age group from 26 to 35, you'll notice that they have a reduced default rate.

When payment is delayed for more than two months, the likelihood of default for the first month  is greater than 50%.

Those who are Divorced or other have higher chance of getting default.

Users with credit limits more than 750k have a 0% likelihood of defaulting.
Female users who are married have the most defaults, while male users who are unmarried or single have the most defaults.
During the first pay month, university graduates have the highest percentage or likelihood of defaulting.
